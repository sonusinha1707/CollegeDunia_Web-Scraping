```python
# Importing Libraries
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import re
```

```python
# Initialize WebDriver
driver = webdriver.Chrome()
# Open a webpage
driver.get("https://collegedunia.com/bsc-colleges")
```

```python
 # Wait for the element with ID "main" to become visible
element = WebDriverWait(driver, 20).until(
    EC.visibility_of_element_located((By.TAG_NAME, "body"))
)
```

```python
# Load all that from Page 
last_height = driver.execute_script("return document.body.scrollHeight")
try:
    
    while True:
        # Scroll down to the bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        
        # Wait to load the page
        time.sleep(2)  # Adjust the sleep time based on the page's loading speed
        
        # Calculate new scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        
        # Check if we've reached the bottom
        if new_height == last_height:
            break
        last_height = new_height

except :
    print("Error occurred while scrolling:")
```

```python
# Fetch the page source after loading with Selenium
soup = BeautifulSoup(driver.page_source,"html.parser")
```

```python
# targeting all the table rows that contain required data
data = soup.find_all("tr",{"class":["jsx-3230181281 table-row even","jsx-3230181281 table-row odd"]})
```

```python
# Creating dictonary to store scrapped data
skeleton = {
    "College Name":[],
    "Location":[],
    "Accreditation":[],
    "CD Rank":[],
    "Courses":[],
    "B.sc Fees":[],
    "Review":[],
    "Rating":[],
    "URL":[]
}
```

```python
# Iterating loop on target table rows
for i in data:
    try:
        print(i.find("td",{"class":"jsx-3230181281 font-weight-medium text-lg position-relative"}).text)
        skeleton["CD Rank"].append(i.find("td",{"class":"jsx-3230181281 font-weight-medium text-lg position-relative"}).text)
    except:
        skeleton["CD Rank"].append("Null")
    try:
        print(i.find("h3",{"class":"jsx-3230181281 font-weight-medium text-lg mb-0"}).text)
        skeleton["College Name"].append(i.find("h3",{"class":"jsx-3230181281 font-weight-medium text-lg mb-0"}).text)
    except:
        skeleton["College Name"].append("Null")
    try:
        print(i.find("span",{"class":"jsx-3230181281 pr-1 location"}).text)
        skeleton["Location"].append(i.find("span",{"class":"jsx-3230181281 pr-1 location"}).text)
    except:
        skeleton["Location"].append("Null")
    try:
        print(i.find("span",{"class":"jsx-3230181281 approvals pl-1"}).text)
        skeleton["Accreditation"].append(i.find("span",{"class":"jsx-3230181281 approvals pl-1"}).text)
    except:
        skeleton["Accreditation"].append("Null")
    try:
        print(i.find("span",{"class":"jsx-3230181281 course-name"}).text)
        skeleton["Courses"].append(i.find("span",{"class":"jsx-3230181281 course-name"}).text)
    except:
        skeleton["Courses"].append("B.Sc")
    try:
        print(i.find("span",{"class":"jsx-3230181281 text-lg text-green d-block font-weight-bold mb-1"}).text)
        skeleton["B.sc Fees"].append(i.find("span",{"class":"jsx-3230181281 text-lg text-green d-block font-weight-bold mb-1"}).text)
    except:
        skeleton["B.sc Fees"].append("Null")
    try:
        print(re.search(r'\d+',i.find("span",{"class":"jsx-3230181281 lr-value d-block font-weight-medium text-sm"}).text).group())
        skeleton["Review"].append(re.search(r'\d+',i.find("span",{"class":"jsx-3230181281 lr-value d-block font-weight-medium text-sm"}).text).group())
    except:
        skeleton["Review"].append("Null")
    try:
        print(re.search(r'\d+',i.find("span",{"class":"jsx-3230181281 lr-key text-lg text-primary d-block font-weight-medium mb-1"}).text).group())
        skeleton["Rating"].append(re.search(r'\d+',i.find("span",{"class":"jsx-3230181281 lr-key text-lg text-primary d-block font-weight-medium mb-1"}).text).group())
    except:
        skeleton["Rating"].append("Null")
    try:
        print(i.find("a",{"class":"jsx-3230181281 college_name underline-on-hover"}).get("href"))
        skeleton["URL"].append(i.find("a",{"class":"jsx-3230181281 college_name underline-on-hover"}).get("href"))
    except:
        skeleton["URL"].append("Null")
    print("___________________________________________________________________")
```

```python
#  create dataframe for scrapped data
import pandas as pd
df = pd.DataFrame(skeleton)
```

```python
# getting idea about null values, number of rows and columns as well as data type
df.info()
```

```python
# download CSV file of dataframe
df.to_csv("collegedunia.csv")
```
